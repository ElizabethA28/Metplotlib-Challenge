```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import random
```

```python
# Study data files
mouse_metadata_path = "Resources/Mouse_metadata.csv"
study_results_path = "Resources/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
# Combine the data into a single DataFrame
pharma_combine = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="left")

# Display the data table for preview
pharma_combine
```

```python
# Checking the number of mice.
mouse_data = pharma_combine["Mouse ID"].unique()
number_of_mice =len(mouse_data)
number_of_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicates = pharma_combine[pharma_combine[["Mouse ID", "Timepoint"]].duplicated()]
mouse_duplicates = pharma_combine["Mouse ID"].unique()
mouse_duplicates
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = pharma_combine.drop_duplicates(subset=["Mouse ID", "Timepoint"])
clean_data
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mouse_data = clean_data["Mouse ID"].unique()
number_of_mice_clean =len(clean_mouse_data)
number_of_mice_clean
```

Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.

# set the index first
drug_regimen = clean_data.set_index(["Drug Regimen"])

# groupby
group_drug_data = drug_regimen.groupby("Drug Regimen")

# mean, median, variance, standard deviation, and SEM of the tumor volume.
mean_drug_regimen = group_drug_data["Tumor Volume (mm3)"].mean()
median_drug_regimen = group_drug_data["Tumor Volume (mm3)"].median()
var_drug_regimen = group_drug_data["Tumor Volume (mm3)"].var()
std_drug_regimen = group_drug_data["Tumor Volume (mm3)"].std()
sem_drug_regimen = group_drug_data["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary DataFrame.
drug_regimen_data = pd.DataFrame ({
    "Mean Tumor Volumn (mm3)": mean_drug_regimen,
    "Median Tumor Volumn (mm3)": median_drug_regimen,
    "Variance Tumor Volumn (mm3)": var_drug_regimen,
    "Standard Deviation Tumor Volumn (mm3)": std_drug_regimen,
    "SEM Tumor Volumn (mm3)": sem_drug_regimen,})
    
drug_regimen_data
```
